```python
import requests
import bs4
import re
import random
import time
import pandas as pd
import csv
```

```python
with open('titlelinks_2007to2017.txt', 'r') as f:
    reader = csv.reader(f)
    link = list(reader)
    link_list = link[1:]
```

```python
review_link =[]
for link in link_list:
    url = 'http://www.imdb.com'+ str(link[0][:17]) + 'reviews?ref_=tt_urv'
    review_link.append(url)
```

```python
# only get review from 100 movies
rand_smpl = [ review_link[i] for i in sorted(random.sample(range(len(review_link)), 1000)) ]
```

```python
review = []
for u in rand_smpl:
    m_url = u
    normal_delay = random.normalvariate(1, 0.2)
    print ("Complete")
    soup = bs4.BeautifulSoup(requests.get(m_url).text,'html5lib')
    data_div = soup.find('div', attrs={'id':'main'})
#     data_html = data_div.get_attribute('innerHTML')
#     soup = bs4.BeautifulSoup(data_html,'html5lib')
    title = data_div.find('h3', attrs = {'itemprop':'name'}).a.text
    lister = soup.find('div', attrs={'class':'lister-list'})
    review_list = lister.find_all('div', attrs={'class':'lister-item'})
    for r in review_list:
        if r.find('span', attrs={'class':'rating-other-user-rating'}):
            rate = r.find('span', attrs={'class':'rating-other-user-rating'}).text
        review_title = r.find('div', attrs={'class':'title'}).text.strip()
        content = r.find('div', attrs={'class':'text'}).text.strip()
        action = r.find('div', attrs={'class':'actions'}).text.strip()
        review_detail = [title, review_title, rate, content, action]
        review.append(review_detail)
    time.sleep(normal_delay)
```

```python
df = pd.DataFrame(review)
```

```python
df.to_csv('reviews_update.csv')
```
